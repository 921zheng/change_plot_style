```python
#make random database
import numpy as np
x=np.random.randn(10,4)
import pandas as pd
df=pd.DataFrame(x, columns=["a","b","c","d"])
df
ax=df.plot(kind="bar")
ax.set(title="random number plot", xlabel="xlabel",ylabel="ylabel")
ax.legend().set_visible(True)
```

```python

df=pd.read_csv("/content/heart-disease.csv")
over_50=df[df["age"]>50]
over_50
```

```python


#add some change
plt.style.use('seaborn-v0_8')
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,6))
scatter=ax0.scatter(x=over_50["age"],y=over_50["trestbps"],c=over_50["target"],cmap="winter")
ax0.set_xlim([50,80])
ax0.set_ylim([60,250])
ax0.set(title="Heart disease and trestbps",xlabel="Age", ylabel="Trestbps")
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(over_50["trestbps"].mean(),color='r', linestyle="--")

scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap="summer")
ax1.set_xlim([50,80])
ax1.set_ylim([60,250])
ax1.set(xlabel="Age",ylabel="Thalach", title="Heart disease and thalach")
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(over_50["thalach"].mean(),color='r', linestyle="--")

#title the figure
fig.suptitle("Heart Disease Analysis")
```
